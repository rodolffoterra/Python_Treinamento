```python
# Load the Package

import pandas as pd
```

```python
# Load the data

df = pd.read_csv("https://raw.githubusercontent.com/dsacademybr/Datasets/master/dataset6.csv")
```

```python
df. head()
```

```python
# Then we extract the fashion from the Quantity columns

moda = df['Quantity'].value_counts().index[0]
moda
```

```python
# And finally we fill NA values with fashion

df['Quantity'].fillna(value = moda, inplace = True)
```

```python
# SDales Column minimum and maximum values 

df.Sales.describe()
```

```python
# Return all sales between 10000 and 12000
df2 = df.query('10000 < Sales < 12000')

df2.Sales.describe()
```

```python
# We apply group By in dataset

df_agg = df[['Region','Sales', 'Quantity']]. groupby(['Region']).agg({'Sales':'mean','Quantity':'sum'})
```

```python
df_agg.head()
```

```python
# We renamed the columns

df_agg.rename(columns = {'Sales':'Media de Vendas','Quantity':"Total Vendida"}, inplace = True)
```

```python
# No we create the rank column which is the result of the rank method that will number items from highest to lowest

df_agg['rank'] = df_agg['Media de Vendas'].rank(method ='first',ascending = False).astype('int')
```

```python
df_agg
```

```python
# Since we want the two highest sales averages, we want items of rank 1 and 2, or whichever is less tham 3

df_agg[df_agg['rank']<3]
```

```python
df_agg[df_agg['Media de Vendas'] > 220]
```

```python
# Now we highlight the highest values in each column in green (So we use axis = 0) and the lowest values

df_agg.style.highlight_max(axis = 0, color = 'green').highlight_min(axis = 0, color = 'blue')
```

```python
# All row where the Qunatity column is equal to the values 5,7,9 or 11

df[df['Quantity'].isin([5,7,9,11])]
```

```python
# Return only 10 lines
df[df['Quantity'].isin([5,7,9,11])][:10]
```

```python
# Filtering the sales that occurred for the Home Office segment and in the South region

df[(df.Segment =='Home Office') & (df.Region =='South')]
```

```python
# Filtering the sales that occurred for the Home Office segment or in the South region

df[(df.Segment =='Home Office') | (df.Region =='South')]
```

```python
# Filtering out sales that did not occur for the Home Office segment or in the South region

df[(df.Segment != 'Home Office') & (df.Region != 'South')]
```

```python
# We filter the dataframe by the Segment column with values that start with the letters 'Con'

df[df.Segment.str.startswith('Con')].head()
```

```python
# We filter the dataframe by the Segment column with values that end with the letters 'ate'

df[df.Segment.str.endswith('ate')].head()
```

```python
df['Customer ID'].head()
```

```python
## We replaced the CG characters with AX in the Customer Id column

df['Customer ID'] = df['Customer ID'].str.replace('CG', 'AX')
df['Customer ID'].head()
```

```python
# We will remove the digits 2 and 0 to the left of the value of the variable 'Ship Date'

df['Ship Date'].str.lstrip('20')

# We can also use the functions rstrip() and strip() with different variations of strip strings
```

```python
df['Order ID'].head()
```

```python
# Sploit the column by the '-' character
df['Order ID'].str.split('-')
```

```python
# We split the column and extract the item in position 2

df['pais'] = df['Order ID'].str.split('-').str[0]

df['ano'] = df['Order ID'].str.split('-').str[1]

df['cod'] = df['Order ID'].str.split('-').str[2]
```

```python
df['pais'].head()
```

```python
df['ano'].head()
```

```python
df['cod'].head()
```

```python
# Concatenating String

df['Nova_Coluna'] = df['Ship Mode'].str.cat(df['Segment'], sep = '-')

df['Nova_Coluna'].head()
```

```python
# Then we define the options, like maximum width of each column, masimun of line,
# maimum of columns and precision of decimal values

pd.set_option('max_colwidth',10)
pd.set_option('max_rows', 20)
pd.set_option('max_columns', 10)
pd.set_option('precision', 2)
pd.options.display.float_format = '{:,.3f}'.format
```
